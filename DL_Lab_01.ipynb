```python
import numpy as np

def mcp_neuron(inputs, weights, threshold):
    summation = np.dot(inputs, weights)
    return 1 if summation >= threshold else 0

weights = np.array([1, 1])
threshold = 2

print("MCP AND Gate")
for inp in [(0,0), (0,1), (1,0), (1,1)]:
    print(inp, "->", mcp_neuron(inp, weights, threshold))
```
